```python
class Rectangle:
    def __init__(self, width=1, height=1):
        self.width = width
        self.height = height
    def getArea(self):
        return self.width * self.height
    def getPerimeter(self):
        return 2 * (self.width + self.height)
    
rect1 = Rectangle(4, 40)
rect2 = Rectangle(3.5, 35.9)
print(f"Rectangle 1 - Width: {rect1.width}, Height: {rect1.height}, Area: {rect1.getArea()}, Perimeter: {rect1.getPerimeter()}")
print(f"Rectangle 2 - Width: {rect2.width}, Height: {rect2.height}, Area: {rect2.getArea():.2f}, Perimeter: {rect2.getPerimeter()}")
```

```python
class Car:
    def __init__(self, door, price, color, brand):
        self.door = door
        self.price = price
        self.color = color
        self.brand = brand
    def getDoor(self):
        return self.door
    def getPrice(self):
        return self.price
    def getColor(self):
        return self.color
    def getBrand(self):
        return self.brand
    
    def setDoor(self, door):
        self.door = door
    def setPrice(self, price):
        self.price = price
    def setColor(self, color):
        self.color = color
    def setBrand(self, brand):
        self.brand = brand
    
    def srartCar(self):
        print("The car has started.")
    def stopCar(self):
        print("The car has stopped.")

mycar = Car(5, 20000, "White", "Toy Car")
print(f"My car is a {mycar.getColor()} {mycar.getBrand()} with {mycar.getDoor()} doors, priced at ${mycar.getPrice()}.")
mycar.srartCar()
mycar.stopCar()

mycar.setColor("Red")
mycar.setPrice(25000)
mycar.setDoor(2)
mycar.setBrand("Ford")
print(f"My car is a {mycar.getColor()} {mycar.getBrand()} with {mycar.getDoor()} doors, priced at ${mycar.getPrice()}.")
mycar.srartCar()
mycar.stopCar()
```
