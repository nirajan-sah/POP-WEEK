```python
while True:
    previous_Meter_Reading = int(input("\nEnter previous meter reading: "))
    present_Meter_Reading = int(input("Enter present meter reading: "))
    day = int(input("Enter day of reading: "))
    month = int(input("Enter month of reading: "))
    print("\nInput Data:", previous_Meter_Reading, present_Meter_Reading, day, month)
    if previous_Meter_Reading < 0 or previous_Meter_Reading > 9999:
        print("Error: previous meter reading out of range")
    if present_Meter_Reading < 0 or present_Meter_Reading > 9999:
        print("Error: present meter reading out of range")
    if previous_Meter_Reading > present_Meter_Reading:
        print("Error: previous reading is more than present reading")
    unitsUsed = present_Meter_Reading - previous_Meter_Reading
    if unitsUsed > 1000:
        print("Error: electricity used exceeds 1000 units")
    if month < 1 or month > 12:
        print("Error: month out of range")
    daysInMonth = {1:31, 2:29, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
    if month in daysInMonth and (day < 1 or day > daysInMonth[month]):
        print("Error: day is invalid for the month")
    continue_input = int(input("Do you want to enter another reading? (0 to finish program): "))
    if continue_input == 0:
        break
```

```python
suspectProfile = []
criminalProfile = []
print("Enter suspect DNA profile (10 numbers):")
for i in range(10):
    a = input(f"Element {i+1}: ")
    suspectProfile.append(a)
    print(a, end=" ")
print("\nEnter criminal DNA profile (10 numbers):")
for i in range(10):
    b = input(f"Element {i+1}: ")
    criminalProfile.append(b)
    print(b, end=" ")
if suspectProfile == criminalProfile:
    print("\nProfile matches")
else:
    print("\nProfiles do NOT match")
```

```python
def matchingProfiles(profile1, profile2):
    return profile1 == profile2

suspectProfile = []
criminalProfile = []
print("Enter suspect DNA profile (10 numbers):")
for i in range(10):
    a = input(f"Element {i+1}: ")
    suspectProfile.append(a)
    print(a, end=" ")
print("\nEnter criminal DNA profile (10 numbers):")
for i in range(10):
    b = input(f"Element {i+1}: ")
    criminalProfile.append(b)
    print(b, end=" ")

if matchingProfiles(suspectProfile, criminalProfile):
    print("\nProfiles match: Suspect is a repeated offender")
else:
    print("\nProfiles do NOT match")
```
